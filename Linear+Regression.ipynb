```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
x= np.array([0,1,2,3,4,5])
y= np.array([0,0.8,0.9,0.6,0.4, 0.8])
```

```python
print(x)
```

```python
print(y)
```

```python
from scipy.interpolate import *
```

```python
p1 = np.polyfit(x,y,1)
```

```python
print(p1)
```

```python
plt.plot(x,y,'o')
```

```python
plt.plot(x,y,'o')
plt.plot(x,np.polyval(p1,x), 'r-')
```

```python
p2 = np.polyfit(x,y,2)
p3 = np.polyfit(x,y,3)
```

```python
print(p1)
```

```python
print(p2)
```

```python
print(p3)
```

```python
plt.plot(x,y,'o')
plt.plot(x,np.polyval(p1,x), 'r-')
plt.plot(x,np.polyval(p2,x), 'b--')
plt.plot(x,np.polyval(p3,x), 'm:')
```

```python
plt.plot(x,y,'o')
xp = np.linspace(-2,6,100)
plt.plot(xp,np.polyval(p1,xp), 'r-')
plt.plot(xp,np.polyval(p2,xp), 'b--')
plt.plot(xp,np.polyval(p3,xp), 'm:')
```

```python
yfit = p1[0] * x +p1[1]
print(yfit)
print(y)
```

```python
yresid = y - yfit
SSresid = sum(pow(yresid, 2))
SStotal = len(y) * var(y)
rsq = 1 - SSresid/SStotal
print(rsq)
```

```python
from scipy.stats import *
slope, intercept, r_value, p_value, std_err = linregress(x, y)
print(pow(r_value,2))
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2,3,4,5,6])
y = np.array([0.2,1.1,1.3,1.2,2.1,1.9,2.5])
```

```python
plt.plot(x,y, 'o')
```

```python
coefint1 =np.polyfit(x,y, 1)
coefint2 =np.polyfit(x,y, 2)
coefint3 =np.polyfit(x,y, 3)
coefint4 =np.polyfit(x,y, 4)

print(coefint1)
print(coefint2)
print(coefint3)
print(coefint4)
```

```python
plt.plot(x,y,'o')
plt.plot(x,np.polyval(coefint1, x),'black')
plt.plot(x,np.polyval(coefint2, x),'g')
plt.plot(x,np.polyval(coefint3, x),'r')
plt.plot(x,np.polyval(coefint4, x),'y')
```

```python
for i in range(1,5):
    print('Degree is ', i)
    p = np.polyfit(x,y,i)
    plt.plot(x,np.polyval(p,x), color='blue')
    plt.show()
```
